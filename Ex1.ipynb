```python
# Find the average of a list of numbers

arr = [1, 2, 3, 4, 5, 6, 7, 8]
print("Array is: ")
for x in arr:
    print(x)

sum = 0
for x in arr:
    sum += x
avg = sum / len(arr)
print("Sum is: ", sum)
print("Average is: ", avg)
```

```python
word = str(input("Enter a word: "))
count = 0
vowel = 'aeiouAEIOU'

for x in word:
    for y in vowel:
        if x == y:
            count += 1
    
print("Number of vowels in the word is: ", count)
```
